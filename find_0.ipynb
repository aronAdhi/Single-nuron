```python
import time
from IPython.display import clear_output
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def func(x):
  return np.sin(x)
```

```python
def d_func(x):
  return np.cos(x)
```

```python
X = 2
res = 0.1
xmin = -5
xmax = 5
ymin = -3
ymax = 3
iter = 10
s_time = 0.5
points = []
```

```python
def set_scene():
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)
```

```python
def plot_function():
  x = np.arange(xmin,xmax,res)
  plt.plot(x,func(x))
```

```python
def plot_tangent():
  global X

  x = np.arange(X-3, X+3,res)
  x_p = X
  y_p = func(x_p)
  points.append([x_p,y_p])
  m = d_func(x_p)
  y = m*x - m*x_p + y_p
  plt.plot(x,y)
  if(m == 0):
    return 1
  step = m*(-0.1)
  # X = y_p/m
  X += step
  print("X = ", X)
  print("slope = ",m)
  return 0
```

```python
def plot_points():
  for x,y in points:
    plt.scatter(x,y,color='orange',s=10)
```

```python
def draw(frame_no):
  set_scene();
  plot_function();
  is_zero = plot_tangent()
  plot_points()
  plt.show()
  return is_zero
```

```python
for frame_no in range(iter):
    clear_output(wait=True)
    is_zero = draw(frame_no)
    if(is_zero):
      break;
    time.sleep(s_time)
```
