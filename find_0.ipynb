```python
import time
from IPython.display import clear_output
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(x):
  return (x + 0.5)**2
```

```python
def d_func(x):
  return 2*x + 1
```

```python
def set_scene():
  plt.xlim(-10,10)
  plt.ylim(-5,10)
```

```python
def plot_function():
  x = np.arange(-10,10,0.1)
  plt.plot(x,func(x))
```

```python
def plot_tangent(frame_no):
  x = np.arange(frame_no-3, frame_no+3,0.1)
  x_p = frame_no
  y_p = func(x_p)
  m = d_func(x_p)
  y = m*x - m*x_p + y_p
  plt.plot(x,y)
  if(m == 0):
    return 1
  return 0
```

```python
def draw(frame_no):
  set_scene();
  plot_function();
  is_zero = plot_tangent((frame_no-25)*0.1)
  plt.show()
  return is_zero
```

```python
for frame_no in range(50):
    clear_output(wait=True)
    is_zero = draw(frame_no)
    if(is_zero):
      break;
    time.sleep(0.5)
```
