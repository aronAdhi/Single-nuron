```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.random.randint(0, 10, 200)
arr2 = arr*2 + 2
#print(arr)
#print(arr2)
```

```python
arr = arr.reshape(-1,1)
arr2 = arr2.reshape(-1,1)
data = np.c_[arr,arr2]
```

```python
print(data[:5])
```

```python
weight = 0
bias = 0
learning_rate = 0.00001
weight_arr = []
bias_arr = []
error_arr = []

for input, actual_output in data:
  predicted_output = input * weight + bias
  error = actual_output - predicted_output
  weight_arr.append(weight)
  bias_arr.append(bias)
  error_arr.append(error)
  weight += (error * input * learning_rate)
  bias += error * learning_rate
```

```python
print(weight)
print(bias)
```

```python
range = 20
plt.plot(np.array(weight_arr[:range]))
plt.show()
plt.plot(np.array(bias_arr[:range]))
plt.show()
plt.plot(np.array(error_arr[:range]))
plt.show()
```
